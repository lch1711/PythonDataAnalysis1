```python
def stock_price(ticker):
    import requests
    from bs4 import BeautifulSoup as bts
    import json
    import pandas as pd
    
    url = f'https://finance.naver.com/item/main.naver?code={ticker}'
    res = requests.get(url)
    soup = bts(markup = res.text, features = 'html.parser')
    items = soup.select('p.no_today > em > span')
    return int(items[0].text.replace(',', ''))
```

```python
stock_price(ticker = '066570')
```

```python
stock_price(ticker = '005930')
```

```python
tickers = ['005930', '373220', '000660', '207940']
```

```python
compNms = ['삼성전자', '엘지엔솔', '하이닉스', '삼바로직']
```

```python
for ticker, compNm in zip(tickers, compNms):
    price = stock_price(ticker)
    print(f'{compNm}의 현재 가격은 {price}원입니다.')
```
