### xlsx 파일 읽기

```python
# xlsx 파일명을 재사용할 수 있도록 변수에 할당합니다.

fileName = 'APT_Price_Seoul_2020_2022.xlsx'
```

### xlsx 파일 입출력

```python
# xlsx 파일의 모든 시트를 읽고 데이터프레임이 값인 딕셔너리를 반환합니다.
# sheet_name : 읽을 시트의 정수 인덱스 또는 시트명을 지정한다. (생략하면 첫 번째 시트를 읽는다. 기본값:0)
# usecols : 사용할 열의 정수 인덱스 또는 열이름을 리스트 같은 반복 가능한 객체로 지정한다. (기본값 : None)
# skiprows : 열이름 위로 생략할 행 개수를 지정한다. (기본값: None)
# thousands : 천 단위 구분자를 문자열로 지정한다. (참고, 콤마가 있는 숫자를 문자로 읽는다.)

pd.read_excel(io = fileName, 
              sheet_name = '2022년', 
              usecols = range(1, 12), 
              skiprows = 3, 
              thousands = ',')
```

```python
%%time
# %%time은 주석 앞에 쓰도록 한다. 
# xlsx 파일의 모든 시트를 읽고 데이터프레임이 값인 딕셔너리를 반환합니다.
# [참고] sheet_name 매개변수에 None을 지정하면 xlsx파일의 모든 시트를 읽고 딕셔너리로 반환한다. 

dfs = pd.read_excel(io = fileName, 
                    sheet_name = None, 
                    usecols = range(1, 12), 
                    skiprows = 3, 
                    thousands = ',')
```

```python
# dfs를 출력합니다.

dfs
```

```python
# dfs의 type을 확인한다.
# [참고] dfs는 키가 시트명이고 값이 데이터프레임인 딕셔너리입니다.

type(dfs)
```

```python
# dfs의 키가 sheet 명이고, 값이 데이터프레임인 딕셔너리이다.

dfs.keys()
```

#### xlsx 파일 입출력 (계속)

```python
# 최종 결과를 저장할 빈 데이터프레임을 생성합니다.

df1 = pd.DataFrame()

# 반복문으로 dfs의 값(데이터프레임)을 행(세로) 방향으로 결합합니다. (axis = 0의 방향)
# df1 : 위 df, dfs[key] : 아래 df, ignore_index = True : 기존 index는 무시한다.

for key in dfs.keys():
    df1 = pd.concat(objs = [df1, dfs[key]], ignore_index = True)
```

```python
# df1의 정보를 확인합니다.
# df1의 행의 개수 139919, 열의 개수 11, 거래일의 자료형은 numpy.datetime64[ns]다.
# [참고] xlsx 파일은 날짜시간 자료형의 셀서식을 유지한다. 

df1.info()
```

```python
# df1을 xlsx 파일로 저장합니다. (read 시간보다 write 시간이 더 오래 소요 됨)
# [참고] index 매개변수에 False를 지정하면 인덱스를 생략합니다.

%time df1.to_excel(excel_writer = 'test.xlsx', index = False)
```

### csv 파일 읽기

```python
# 현재 작업 경로에 있는 폴더명과 파일명을 files에 할당합니다.

files = os.listdir()
files
```

```python
# for 반복문으로 files (os.listdir) 에 'csv' 와 'Price'가 포함된 파일을 반환한다.

for file in files :
    if 'csv' in file and 'Price' in file:
        print(file)
```

```python
# 'csv'와 'Price'를 함께 포함하는 파일명을 fileNames에 할당합니다.

fileNames = [file for file in files if 'csv' in file and 'Price' in file]
fileNames
```

### csv 파일의 문자 인코딩 방식 확인

```python
# csv 파일을 bytes로 읽고 text에 할당합니다.
# [참고] with 문에서 나오면서 close()를 호출합니다.
# mode (w : 쓰기, a : 추가, r : 읽기, rb : 바이너리(byte))

with open(file = fileNames[0], mode = 'rb') as file:
    text = file.read()
```

```python
# text의 일부를 출력합니다.
# [참고] text는 ASCII로 인코딩된 문자열이므로 사람이 읽기 어렵습니다.
# [참고] 따옴표 왼쪽에 b가 보이면 bytes 자료형입니다.

text[:100]
```

```python
# text의 일부에 사용된 문자 인코딩 방식과 신뢰도를 확인합니다.
# [주의] 지정한 문자열에 따라 결과가 달라질 수 있습니다.

chardet.detect(text[:100])
```

### csv 파일 입출력

```python
%%time

# 최종 결과를 저장할 빈 데이터프레임을 생성합니다.

df2 = pd.DataFrame()

# 반복문으로 csv 파일을 읽고 데이터프레임을 행(세로) 방향으로 결합합니다.

for fileName in fileNames:
    df = pd.read_csv(filepath_or_buffer = fileName, 
                     encoding = 'CP949')
    df2 = pd.concat(objs = [df2, df], ignore_index = True)
```

```python
# df2의 정보를 확인합니다.
# [참고] parse_dates 매개변수를 추가하지 않으면 거래일의 자료형을 object로 변환한다.
# [참고] thousands 매개변수를 추가하면 거래금액을 정수형으로 변환한다. 

df2.info()
```

```python
%%time
df1 = pd.DataFrame()
for fileName in fileNames:
    df = pd.read_csv(fileName, 
                     encoding = 'CP949', 
                     parse_dates = ['거래일'], 
                     thousands = ',')
    df1 = pd.concat(objs = [df1, df], ignore_index = True)
```

```python
df1.info()
```

```python
# df2를 csv 파일로 저장합니다.
# \ : 칸 변경 후 유지하라는 의미로 스페이스가 없어야 한다. 

%time df2.to_csv(path_or_buf = 'test.csv', \
                 encoding = 'UTF-8', \
                 index = False)
```
